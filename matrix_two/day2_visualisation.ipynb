```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_hdf('../data/car.h5')
df.shape
```

```python
df['price_value'].hist(bins=100)
```

```python
def group_and_barplot(feat_groupby, feat_agg='price_value', agg_funcs=[np.mean, np.median, np.size], by='mean', top=50, subplots=True):
    return (
        df
        .groupby(feat_groupby)[feat_agg]
        .agg(agg_funcs)
        .sort_values(by=by, ascending=False)
        .head(top)
    ).plot(kind='bar', figsize=(15,5), subplots=subplots)
```

```python
group_and_barplot('param_marka-pojazdu');
```

```python
group_and_barplot('param_kraj-pochodzenia', by='size');
```

```python
group_and_barplot('param_kolor', by='mean');
```
